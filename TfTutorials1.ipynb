```
def func1(x,y):
  if x>10:
    print("x greater than 10")
  elif y<10:
    print("y less than 10")
  else:
    print("ouside x and y")
  return x+y

func1(10,10)
```

```
for i in range(5,25,5):
  print("running with i {}".format(i))
  r = func1(i,i)
  print(r)
  print("Result of func {} {}".format(str(i),str(r)))
```

```
#Numpy
import numpy as np
a = np.array(["Hello", "world"])
a = np.append(a,"!")
print("Current array {}".format(a))
for i in a:
  print(i)
for i,e in enumerate(a):
  print("{} {}".format(i,e))
```
